```python
import pandas as pd
```

```python
df = pd.read_csv("popular_shows_dataset.csv")
pd.set_option("display.max_row",None)
pd.set_option("display.max_column",None)
df.head()
```

```python
df.info()
df.isna().sum()
```

```python
df.shape
```

```python
df.describe()
```

```python
df["title"] = df["title"].astype(str).str.lower().str.replace('"', '', regex=False).str.replace('!', '', regex=False)
df["title"] = df["title"].str.title().str.replace(r"\bIi\b", "II", regex=True)
df["title"].head()
```

```python
#cleaning title column

df["title"].value_counts(dropna=False)
```

```python
#df.nunique()
```

```python
#clean year column 

df["year"].value_counts(dropna=False)
```

```python
df["revenue"] = df["revenue"].astype(float)
```

```python
df["revenue"].value_counts(dropna=False)
```

```python
df["revenue"] = df["revenue"].astype(str).str.strip()
df["revenue"] = pd.to_numeric(df["revenue"],errors = "coerce")
df["revenue"].head()
```

```python
df["imdb_id"].value_counts(dropna=False)
```

```python
df["imdb_id"] = df["imdb_id"].astype(str).str.extract(r"(\d+)")
df["imdb_id"] = "tt" + df["imdb_id"].str.zfill(9)
df["imdb_id"].head()
```

```python
df["tmdb_id"].value_counts(dropna=False)
```

```python
df["tmdb_id"] = df["tmdb_id"].astype(str).str.strip().str.extract(r"(\d+)")
df["tmdb_id"] = df["tmdb_id"].str.zfill(10)
df["tmdb_id"].head()
```

```python
df["trakt_id"].value_counts(dropna=False)
```

```python
df["trakt_id"] = df["trakt_id"].astype(str).str.strip().str.extract(r"(\d+)")
df["trakt_id"] = df["trakt_id"].str.zfill(11)
df["trakt_id"].head()
```

```python
df.shape
```

```python
df.isna().sum()
```

```python
df.duplicated().sum()
```

```python
df.value_counts().head(10)
```

```python
df = df.drop_duplicates()
```

```python
df.shape
```
